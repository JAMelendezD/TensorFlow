```python
plt.xkcd()

def f(x):
    return X*X - 2*np.exp(-5*(X-2)*(X-2)) 

xmin = 0
xmax = 2.32

X = np.linspace(xmin,xmax,100)
Y = f(X)

fig = plt.figure(figsize=(12,5), dpi=72,facecolor="white")
axes = plt.subplot(111)

plt.plot(X,Y, color = 'k', linewidth=2, linestyle="-", zorder=+10)

axes.set_xlim(X.min(),X.max())
axes.set_ylim(1.01*Y.min(), 1.01*Y.max())

axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')
axes.xaxis.set_ticks_position('bottom')
axes.spines['bottom'].set_position(('data',0))
axes.yaxis.set_ticks_position('left')
axes.spines['left'].set_position(('data',X.min()))

axes.set_xticks([])
axes.set_yticks([])
axes.set_xlim( 1.05*X.min(), 1.10*X.max() )
axes.set_ylim( 1.15*Y.min(), 1.05*Y.max() )

T = [10,35,50,67,82,99]
plt.scatter( X[T], Y[T], s=50, zorder=+12, c='k')


colors = []

for t in T:
    if X[t] < xmax/2.0:
        new_color = (Y[t]/np.max(Y[T]), 0, X[t]/np.max(X[T]))
        colors.append(new_color)
    elif X[t] >= xmax/2.0:
        new_color = (Y[t]/np.max(Y[T]), 0, X[t]/np.max(X[T]) )
        colors.append(new_color)
    
plt.text(X[T[0]]-.1, Y[T[0]]+.1, "Neural-Net\nNoob", ha='left', va='bottom', color = colors[0])
plt.text(X[T[1]]-.15, Y[T[1]]+.3, "Raspberry", ha='left', va='bottom',color = colors[1])
plt.text(X[T[2]]-0.1, Y[T[2]]+0.3, "Cat\nDog", ha='left', va='center',color = colors[2])
plt.text(X[T[3]]-0.1, Y[T[3]]+0.3, "Overfitting", ha='left', va='center',color = colors[3])
plt.text(X[T[4]]+.05, Y[T[4]], "One GPU\nLimit", ha='left', va='center',color = colors[4])
plt.text(X[T[5]]+.05, Y[T[5]], "VGG16", ha='left', va='center',color = colors[5])
plt.text(X[T[2]]+0.42, 4.0, "Learning\nPlateau", ha='center', va='center', fontsize=24, color = 'k')

plt.ylabel("Expected Result",y=.5, fontsize=20, color = 'k')
plt.text(-.1, 4.3, "+",ha='left', va='top', color=(1,0,0,1), fontsize=40, rotation = 'vertical')
plt.text(.05, -.1, "Computer Power",ha='left', va='top', color='k', fontsize=20)
plt.text(0.84, -.15, ">",ha='left', va='top', color=(0,0,0.5,1), fontsize=20)
plt.text(.774, -.142, "---",ha='left', va='top', color='0.3', fontsize=15)
plt.text(0.75, -.15, "<",ha='left', va='top', color='k', fontsize=20)

X = np.linspace(0,1.1*xmax,100)
axes.fill_between(X, 0, +10, color = (1,1,0.8,0.5), zorder=-1)


X = np.linspace(1.5,2.25,100)
Y = f(X)

axes.axvline(x=1.35,ymin=0,ymax=1, color='.5', ls='--', linewidth = '2')
axes.axvline(x=1.85,ymin=0,ymax=1, color='.5', ls='--', linewidth = '2')

#plt.savefig("xkcd.png", dpi  = 600)
plt.show()
```
